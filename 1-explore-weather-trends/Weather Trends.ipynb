# Project 1: Explore Weather Trends

```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import CSV

globaltemp = pd.read_csv('global_results.csv', index_col='year')
citytemp = pd.read_csv('chicago_results.csv', index_col='year')

globaltemp.head(), citytemp.head()
```

```python
# Calculate Moving Average

globaltemp['sma_5'] = globaltemp.avg_temp.rolling(5).mean()
globaltemp['sma_10'] = globaltemp.avg_temp.rolling(10).mean()

citytemp['sma_5'] = citytemp.avg_temp.rolling(5).mean()
citytemp['sma_10'] = citytemp.avg_temp.rolling(10).mean()

globaltemp.head(10), citytemp.head(10)
```

```python
# Clean Up Data

globaltemp.drop(['sevenDayMA','FiveYearMA','TenYearMA'], axis=1, inplace=True)

citytemp.drop(['city','country','sevenDayMA','FiveYearMA','TenYearMA'], axis=1, inplace=True)

globaltemp.head(10), citytemp.head(10)
```

```python
# Plot Line Chart

plt.figure(figsize = (24,10))

globaltemp['avg_temp'].plot(label='Global Temp Avg', color='orange')
globaltemp['sma_5'].plot(label='Global Temp 5-Year SMA', color='red')
globaltemp['sma_10'].plot(label='Global Temp 10-Year SMA', color='brown')

citytemp['avg_temp'].plot(label='City Temp Avg', color='blue')
citytemp['sma_5'].plot(label='City Temp 5-Year SMA', color='green')
citytemp['sma_10'].plot(label='City Temp 10-Year SMA', color='purple')


plt.title('World vs Chicago Average Temperatures')
plt.xlabel('Year')
plt.ylabel('Temperature (C)')
plt.legend()

plt.show()
```

### Observations

When looking at global average temperatures vs Chicago average temperatures, we can see that Chicago is hotter on average than the global average. Looking at the graph, I can approximate that Chicago's average temperature is 10 Celsius while the global average temperature is 8 Celsius.

When looking at the averages over moving averages, there is  more variance between the years for Chicago's temperatures. For example, around the years 1920-1930, Chicago's temperature fluctuated between 7-12 while the global temperature fluctuated very little and stayed around 8. 

Over time, Chicago has been consistently hotter than the global average. We can see that in general the moving average is warmer by at least 2 Celsius. 

Both have temperatures trending consistently upward. We can see that both averages remain fairly flat between 1875-1975. From 1975 onwards to the 2000s, there has been a significant increased in average temperature.

Overall, we can conclude that the world is getting hotter.
