```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_csv('house_price_area_only.csv')
df.head()
```

```python
df['intercept'] = 1

lm = sm.OLS(df['price'], df[['intercept', 'area']]) # ordinary least squares method, provide y and list of x values
results = lm.fit()

results.summary()
```

```python
# coef - 348 is the slope b1, 9587 is the intercept b0

# p values test of intercept and slope are = 0 in the null hypothesis

# Hypothesis test
# the hypothesis test for each x-variable is a test of if that population slope is equal to zero 
# if the slope is different than zero (the alternative is true), we have evidence that the x-variable attached 
# to that coefficient has a statistically significant linear relationship with the response
# --> suggests that x-variable should help in predicting response 

# R-squared value: 67.8% of variability in price is explained by area of house
```
