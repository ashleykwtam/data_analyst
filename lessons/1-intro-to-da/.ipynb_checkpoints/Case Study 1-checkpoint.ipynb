## Wine Samples Investigation

#### Investigate datasets on chemical properties and quality ratings of wine samples 

Source: https://archive.ics.uci.edu/ml/datasets/Wine+Quality

##### Ask Questions

Given the dataset, I would ask questions like:
- Do the physiochemical properties (acidity, sulfur dioxide, sugar, alcohol, etc) affect the quality rating of the wine?
- What properties are most important to predict wine quality?
- How do the wine types differ in chemical properties? Would wine type affect quality? 
- Do wines with certain higher chemical properties receive higher ratings?
- Do wines with certain lower chemical properties receive higher ratings?

##### Wrangle Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('winequality-red.csv', sep=';')
df2 = pd.read_csv('winequality-white.csv', sep=';')
```

```python
# Understand the dataset

df.head()
df2.head()
```

```python
df.info()
df2.info()
```

```python
# num of samples? num of columns?

print("There are {} samples and {} columns in the red wine dataset".format(df.shape[0], df.shape[1]))
print("There are {} samples and {} columns in the white wine dataset".format(df2.shape[0], df2.shape[1]))
```

```python
# features with missing values?

missing = df.isnull().any()
missing2 = df2.isna().any()

print(missing)
print(missing2)
```

```python
# duplicate rows in the white wine dataset?

df2.duplicated()
```

```python
# number of unique values for quality in each dataset

df['quality'].nunique()
```

```python
# mean density of the red wine dataset

df['density'].mean()
```

```python
# Combine the red and white dataset

# create a colour category for the datasets

red = np.repeat("red",1599)
white = np.repeat("white",4898)

df['colour'] = red
df2['colour'] = white
df2.head()

# append dataframes

wine = df.append(df2, ignore_index=True)
wine.info()

# save as new csv

wine.to_csv('winequality_combined.csv', index=False)

wine.head()
```

##### Explore Data

```python
# Based on histograms of columns in this dataset, which of the following feature variables appear skewed to the right?
# Fixed Acidity, Total Sulfur Dioxide, pH, Alcohol

wine.hist(column=['fixed acidity', 'total sulfur dioxide', 'pH', 'alcohol'], figsize=(15,10));
```

```python
# Based on scatterplots of quality against different feature variables, which of the following is most likely to have a positive impact on quality? 
# Volatile Acidity, Residual Sugar, pH, Alcohol

wine.plot.scatter('quality','volatile acidity', figsize=(6,4)); 
wine.plot.scatter('quality','residual sugar', figsize=(6,4));
wine.plot.scatter('quality','pH', figsize=(6,4));
wine.plot.scatter('quality','alcohol', figsize=(6,4));
```

```python
# Group By

wine.groupby(['quality','colour']).mean() # use as_index = False if you don't want quality and colour to be indexed
```

##### Conclusions

```python
# Is a certain type of wine (red or white) associated with higher quality?

wine.groupby(['colour']).mean()
```

```python
# What level of acidity (pH value) receives the highest average rating?

wine.describe()
```

```python
levels = [2.72, 3.11, 3.21, 3.32, 4.01]

categories = ['High', 'Moderately High', 'Medium', 'Low']

# Add acidity level categories to original dataset
wine['acidity_levels'] = pd.cut(df['pH'], levels, labels=categories)

wine.head()

wine.to_csv('winequality_combined', index=False)
```

```python
wine.groupby(['acidity_levels']).mean()
```

```python
# Do wines with higher alcoholic content receive better ratings?

wine['alcohol'].median()
```

```python
low_alcohol = wine.query('alcohol < 10.3')
high_alcohol = wine.query('alcohol >= 10.3')

print(low_alcohol.mean())
print(high_alcohol.mean())
```

```python
# Do sweeter wines (more residual sugar) receive better ratings?

wine['residual sugar'].median()
```

```python
low_sugar = wine.query('`residual sugar` < 3')
high_sugar = wine.query('`residual sugar` >= 3')

print(low_sugar['quality'].mean())
print(high_sugar['quality'].mean())
```

```python
# Plots

wine.groupby('colour')['quality'].mean().plot(kind='bar', title='Avg Wine Quality by Colour');
plt.xlabel('Wine Colours'); # use matplotlib
plt.ylabel('Quality');
```

```python
# Count each quality by wine colour

colours = ['red', 'black']
counts = wine.groupby(['quality', 'colour']).count()['pH']
counts.plot(kind='bar', title='Count by Wine Colour & Quality', color=colours);
plt.xlabel('Quality and Colour')
plt.ylabel('Count');
```

```python
# There are more whites than reds so it's hard to make a fair comparison

# Proportions

totals = wine.groupby(['colour']).count()['pH']
proportions = counts / totals
proportions.plot(kind='bar', color=colours);
```
